**X has 81 null values in "GarageBltYr" column; while Y has no null values. Hence, we shall be filling in those null values:**

```python
X_tr=X.fillna(2008)
X_tr
```

```python
Y_tr=df.SalePrice
Y_tr
```

**Splitting the dataset in 80:20 as "train" & "test"**:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X_tr, Y_tr, train_size=0.8, test_size=0.2, random_state=0)
```

# **MODELLING WITH RANDOM FOREST REGRESSION:**

```python
# Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
 
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
 
# fit the regressor with x and y data
regressor.fit(X_train,Y_train)
```

**Checking model score in the training data:**

```python
regressor.score(X_train, Y_train)
```

**Predicting outcome on test dataset using the trained model:** 

```python
Y_pred = regressor.predict(X_test)
Y_pred
OUTCOME=pd.DataFrame(Y_pred, columns=["SalePrices"])
OUTCOME
```

```python
OUTCOME.to_csv("OUTCOMES.csv")
```

**model score on test data:**

```python
regressor.score(X_test, Y_pred)
```

**calculating error between the predicted outcome (Y_pred) and actual outcome(Y_test)**

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(Y_test, Y_pred)
```
